```python
def time(duration):
    if duration < 3:
        return 'Short'
    elif 3 <= duration < 4: # in between 3 and 4
        return 'Medium'
    else:
        return 'Long'
```

```python
df['Listening Time'] = df['Avg Stream Duration (Min)'].apply(time)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")
plt.figure(figsize=(18, 28))

# 1. Top 10 Artists by Total Streams
plt.subplot(5, 2, 1)
top_artists =x=top_artists.values, y=top_artists.index, palette="viridis")
plt.title("Top 10 Artists by Total Streams")

# 2. Genre Distribution
plt.subplot(5, 2, 2)
genre_counts = df["Genre"].value_counts().nlargest(10)
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="muted")
plt.title("Top 10 Genres by Number of Entries")

# 3. Country-wise Total Streams
plt.subplot(5, 2, 3)
country_streams = df.groupby("Country")["Total Streams (Millions)"].sum().nlargest(10)
sns.barplot(x=country_streams.values, y=country_streams.index, palette="coolwarm")
plt.title("Top 10 Countries by Total Streams")

# 4. Average Stream Duration by Genre
plt.subplot(5, 2, 4)
avg_duration_genre = df.groupby("Genre")["Avg Stream Duration (Min)"].mean().nlargest(10)
sns.barplot(x=avg_duration_genre.values, y=avg_duration_genre.index, palette="plasma")
plt.title("Average Stream Duration by Genre")

# 5. Top Albums by Monthly Listeners
plt.subplot(5, 2, 5)
top_albums = df.groupby("Album")["Monthly Listeners (Millions)"].sum().nlargest(10)
sns.barplot(x=top_albums.values, y=top_albums.index, palette="cividis")
plt.title("Top 10 Albums by Monthly Listeners")

# 6. Release Year Trends in Total Streams
plt.subplot(5, 2, 6)
yearly_streams = df.groupby("Release Year")["Total Streams (Millions)"].sum()
sns.lineplot(x=yearly_streams.index, y=yearly_streams.values, marker="o", color="teal")
plt.title("Total Streams by Release Year")

# 7. Platform Type vs. Skip Rate
plt.subplot(5, 2, 7)
sns.boxplot(x="Platform Type", y="Skip Rate (%)", data=df, palette="Set2")
plt.title("Platform Type vs. Skip Rate")

# 8. Streams Last 30 Days by Country
plt.subplot(5, 2, 8)
country_last_30 = df.groupby("Country")["Streams Last 30 Days (Millions)"].sum().nlargest(10)
sns.barplot(x=country_last_30.values, y=country_last_30.index, palette="magma")
plt.title("Top 10 Countries by Streams in Last 30 Days")

# 9. Correlation: Monthly Listeners vs. Skip Rate
plt.subplot(5, 2, 9)
sns.scatterplot(x="Monthly Listeners (Millions)", y="Skip Rate (%)", data=df, alpha=0.7)
plt.title("Monthly Listeners vs. Skip Rate")

# 10. Total Hours Streamed by Genre
plt.subplot(5, 2, 10)
genre_hours = df.groupby("Genre")["Total Hours Streamed (Millions)"].sum().nlargest(10)
sns.barplot(x=genre_hours.values, y=genre_hours.index, palette="spring")
plt.title("Total Hours Streamed by Genre")

plt.tight_layout()
plt.show()
```

```python
Adding column  skip rate Category (Low, Medium, High)
```

```python
def skip_rate(rate):
    if rate < 20:
        return 'Low'
    elif 20 <= rate < 50:
        return 'Medium'
    else:
        return 'High'

# Apply the function to create a new column
df["Skip Rate Category"] = df["Skip Rate (%)"].apply(skip_rate)
```
